Perceptron on training data

Student Study Hours-0(Fail)|1(Pass)

```python
import numpy as np
```

```python
X=np.array([1,2,3,4]) #study hours
y=np.array([0,0,1,1]) #fail(0),pass(1)
X,y
```

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial Weight:",w)
print("Initial Bias:",b)
```

Define Activation Function(Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        #1. Compute Weighted sum
        z=w*X[i]+b
        #2. Apply Activation function
        y_pred=step_function(z)
        #3. Compute Error
        error=y[i]-y_pred
        #4.Updated Weight
        w=w + learning_rate*error*X[i]
        #5. Update Bias
        b=b + learning_rate*error
        print(f"Input:{X[i]}, Predicted:{y_pred}, Actual: {y[i]},w:{w},b:{b}")
    
```

```python
print("Final Weight:",w)
print("Final Bias:",b)
```

Testing the Model

```python
for i in range(len(X)):
    z=np.dot(w,X[i])+b
    prediction=step_function(z)
    print("Input",X[i],"Prediction:",prediction)
```

Perceptron OR gate

```python
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,1,1,1])
X,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10
print("Initial Weights:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 1
```

```python
print("\n Testing OR Gate")
for epoch in range(epochs):
    print(f"\Epoch {epoch+1}")
    for i in range(len(X)):
        #1
        z=np.dot(w,X[i])+b
        #2.
        y_pred=step_function(z)
        #3.
        error=y[i]-y_pred
        #4.
        w+=learning_rate*error*X[i]
        #5.
        b+=learning_rate*error
        print(f"Input:{X[i]},Predited:{y_pred}, Actual:{y[i]},w:{w},b:{b}")
```

```python
print("Final Weights:",w)
print("Final Bias:",b)
```

```python
print("\n Testing OR Gate")
for i in range(len(X)):
    z=np.dot(w,X[i])+b
    prediction=step_function(z)
    print("Input",X[i],"Prediction:",prediction)
```
